<font color='red'> Write a script for the calculation of the mean, standard deviation, kurtosis and skewness of Number of follower per user in case of Fake News and Real News dataset. Repeat this process for Favorite Count as well. Conclude whether one can discriminate the two classes using such statistical data.</font> 

```python
# Calculate the mean, standard deviation, kurtosis, and skewness of the number of followers in the fake news dataset

def get_stat_follower_per_user(df, user):
    """This function will calculate the mean, standard deviation, kurtosis, and skewness of the number of followers
    of a user """
    user_data = df.loc[df['user_screen_name'] == user]
    followers_mean = np.mean(user_data['user_followers_count'])
    followers_std = np.std(user_data['user_followers_count'])
    followers_kurtosis = kurtosis(user_data['user_followers_count'])
    followers_skewness = skew(user_data['user_followers_count'])

    return followers_mean, followers_std, followers_kurtosis, followers_skewness
```

```python
#For fake news dataset
dataframe = fake_news
user_name = 'wikileaks'
mean, std, kurt, skewness = get_stat_follower_per_user(dataframe, user_name)

print(f'The mean number of follower of {user_name} is', round(mean,2))
print(f'The standard deviation of number of follower of {user_name} is', round(std,2))
print(f'The Kurtosis of number of follower of {user_name} is', round(kurt,2))
print(f'The skewness of number of follower of {user_name} is', round(skewness,2))
```

```python
#For real news dataset
dataframe = real_news
user_name = 'wikileaks'
mean, std, kurt, skewness = get_stat_follower_per_user(dataframe, user_name)

print(f'The mean number of follower of {user_name} is', round(mean,2))
print(f'The standard deviation of number of follower of {user_name} is', round(std,2))
print(f'The Kurtosis of number of follower of {user_name} is', round(kurt,2))
print(f'The skewness of number of follower of {user_name} is', round(skewness,2))
```

We want to compare the activity of individual users in Fake News and Real News dataset. Select the three most active users in terms of number tweets generated and calculate the average number of tweets generated by the three users. Repeat this process for the five most active users in each dataset, and for the first 10-users in each dataset, and first 15 users for each dataset.
 </font>

```python
# Count the number of tweets generated by top n  user

def compare_top_activity_user(df, number_of_user = 3):
    """this function takes the twitter dataframe and number of required most active users.
    and calculate the average number of tweets generated by the users."""
    tweet_counts = df['user_screen_name'].value_counts()
    # Select the three most active users
    top_users = tweet_counts.index[:number_of_user]
    # Filter the dataframe to only include tweets from the top three users
    top_tweets = df[df['user_screen_name'].isin(top_users)]
    # Calculate the average number of tweets generated by the top three users
    print(f"The {number_of_user} most active users are:")
    print(top_users)
    
    for user in top_users:
        user_tweets = top_tweets[top_tweets['user_screen_name'] == user]
        average_tweets = user_tweets.shape[0] / user_tweets['user_screen_name'].nunique()
        print("The average number of tweets generated by", user, "is:", average_tweets)
    
    average_tweets_top_n = top_tweets.groupby('user_screen_name').size().mean()
    print(f"The average number of tweets generated by top {number_of_user} user is : ", round(average_tweets_top_n,1))

```

```python
# compare the activity of individual users in Fake News of top 3 users.
print('For Fake News dataset: ')
compare_top_activity_user(fake_news)
# compare the activity of individual users in real News of top 3 users.
print('For Real News dataset: ')
compare_top_activity_user(real_news)
```

```python
# compare the activity of individual users in Fake News of top 5 users.
print('For Fake News dataset: ')
compare_top_activity_user(fake_news,5)
# compare the activity of individual users in real News of top 5 users.
print('For Real News dataset: ')
compare_top_activity_user(real_news,5)
```
